```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
aj_plano = cv2.imread("tablero.jpg")
aj_plano = cv2.cvtColor(aj_plano,cv2.COLOR_BGR2RGB)
plt.imshow(aj_plano)
```

```python
aj_plano_gris = cv2.cvtColor(aj_plano,cv2.COLOR_BGR2GRAY)
plt.imshow(aj_plano_gris,cmap='gray')
```

```python
aj_real = cv2.imread("ajedrez_real.jpg")
aj_real = cv2.cvtColor(aj_real,cv2.COLOR_BGR2RGB)
plt.imshow(aj_real)
```

```python
aj_real_gris = cv2.cvtColor(aj_real,cv2.COLOR_BGR2GRAY)
plt.imshow(aj_real_gris,cmap='gray')
```

```python
aj_plano_gris_float = np.float32(aj_plano_gris)
```

```python
aj_plano_gris
```

```python
dst = cv2.cornerHarris(src=aj_plano_gris_float, blockSize=5, ksize=3, k=0.04)

dst = cv2.dilate(dst, None)

aj_plano[dst > 0.1 * dst.max()] = [255, 0, 0]


plt.imshow(aj_plano)
```

```python
aj_real_gris_float = np.float32(aj_real_gris)

dst = cv2.cornerHarris(src=aj_real_gris_float, blockSize=5, ksize=3, k=0.04)

dst = cv2.dilate(dst, None)

aj_real[dst > 0.1 * dst.max()] = [255, 0, 0]

plt.imshow(aj_real)
```

```python
aj_plano = cv2.imread("tablero.jpg")
aj_plano = cv2.cvtColor(aj_plano,cv2.COLOR_BGR2RGB)
plt.imshow(aj_plano)

aj_real = cv2.imread("ajedrez_real.jpg")
aj_real = cv2.cvtColor(aj_real,cv2.COLOR_BGR2RGB)
plt.imshow(aj_real)
```

```python
esquinas = cv2.goodFeaturesToTrack(aj_plano_gris, maxCorners=81, qualityLevel=0.01, minDistance=10 )
esquinas = np.intp(esquinas)
```

```python
esquinas
```

```python
for i in esquinas:
    x,y = i.ravel()
    x,y = int(x), int(y)
    cv2.circle(aj_plano,(x,y),5,255,-1)

plt.imshow(aj_plano)
```

```python
esquinas = cv2.goodFeaturesToTrack(aj_real_gris, maxCorners=200, qualityLevel=0.01, minDistance=10 )
esquinas = np.intp(esquinas)

for i in esquinas:
    x,y = i.ravel()
    x,y = int(x), int(y)
    cv2.circle(aj_real,(x,y),5,255,-1)

plt.imshow(aj_real)
```
