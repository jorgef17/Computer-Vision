```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('perro_cara.jpg')
plt.imshow(img)
```

```python
bordes = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(bordes)
```

```python
bordes = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(bordes)
```

```python
# formula

mediana = np.median(img)
lower = int(max(0,0.7*mediana))
upper = int(min(255,1.3*mediana))

bordes = cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(bordes)
```

```python
blur = cv2.blur(img, ksize=(3,3))
bordes = cv2.Canny(image=blur,threshold1=lower,threshold2=upper)
plt.imshow(bordes)
```
